```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data.csv')
```

```python
df.describe()
```

```python
# Histogram for Maxpulse
sns.histplot(data=df, x='Maxpulse', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Maxpulse')
plt.xlabel('Maxpulse')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for Calories (handle missing values)
sns.histplot(data=df, x='Calories', bins=20, kde=True, color='lightgreen')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```
